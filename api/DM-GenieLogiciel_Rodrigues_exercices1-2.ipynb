```python
# (3) Ajouter des données à la base - pour affichage uniquement, la méthode INSERT via flask n'a pas fonctionné (code commenté à la fin 
# de app/explore_db.py pour référence):

# En tournant cette cellulle j'ajoute une nouvelle phrase à la base de données, en modifiant les tables concernées par le contenu
# du dictionnaire de la nouvelle phrase, donc les tables: 'texte', 'transcription' et celle de joincture. Le code calcule le nouvel
# id suivant la taille maximale de la table de joincture. Les trois tables sont affichées pour montreer l'insertion de l'entrée #21.

import requests


url = 'http://127.0.0.1:5000/api/v1/dialectes/ajoute_phrase'

maphrase = {'texte': 'J’ai tenu à apporter cet objet que j’aime beaucoup.',
     'transcription': 'ZE t@ny a apORte sEtObZE k@ ZEm boku',
     'dialecte': 'français breton'}

x = requests.post(url, json = maphrase)

print(x.text)
```

```python
# (4) Mettre à jour des données dans la base:
# Exemple pour changer le dialecte d'une phrase d'identifiant id = 16 dans la table
# de joincture. classeID passera de 2 (code de "français 'standard'"") à 1 (code de "français breton"):

import requests

url = 'http://127.0.0.1:5000/api/v1/dialectes/update_dialecte_phrase'

maphrase = {'id': 19,
            'texte': 'L’après-midi on faisait de la musique.',
     'transcription': 'lapREmidi O~ fezE d@ la myzik',
     'dialecte': 'français breton'}

x = requests.put(url, json = maphrase)

print(x.text)
```

```python
# (5) Supprimer des données spécifiques de la base:
# Exemple pour supprimer l'entrée dont l'identifiant est id=3 de la table de joincture.

import requests

url = 'http://127.0.0.1:5000/api/v1/dialectes/delete_phrase_jonc'

entree_nb = {'id': 3}

x = requests.get(url, params= entree_nb)

print(x.text)

```
